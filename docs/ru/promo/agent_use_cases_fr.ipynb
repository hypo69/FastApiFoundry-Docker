# Cas d'utilisation de l'agent

## Scène 1
Un agent d'IA local est utile partout où une entreprise travaille avec de vrais documents, des processus internes et des données que l'on ne peut pas simplement envoyer dans le cloud.

## Scène 2
Un service juridique peut téléverser des contrats, des décisions de justice, des correspondances et des expertises. L'agent aide à repérer rapidement les risques, comparer les versions de documents, préparer de courts résumés et répondre aux questions à partir des pièces du dossier.

## Scène 3
Une clinique médicale peut utiliser l'agent pour travailler avec des règlements internes, des protocoles anonymisés, des consignes et des documents de référence. La condition essentielle reste respectée : les données restent à l'intérieur de l'organisation.

## Scène 4
Une équipe financière obtient un assistant pour analyser des rapports, des tableaux, des registres de paiement et des règles internes. L'agent peut expliquer les écarts, retrouver les bons chiffres et préparer une ébauche de commentaire de gestion.

## Scène 5
Les équipes d'ingénierie et d'IT peuvent connecter des logs de serveurs, des manuels, de la documentation et l'historique des incidents. L'agent aide à comprendre ce qui s'est passé, quels systèmes ont été touchés et quelles étapes sont déjà décrites dans les procédures internes.

## Scène 6
Une entreprise industrielle peut créer des assistants pour les procédures, les fiches d'équipement, les consignes de sécurité et les rapports de quart. Un agent travaille avec la documentation, un autre surveille les événements, un troisième aide à former les employés.

## Scène 7
Pour un dirigeant, un tel agent devient un point d'accès unique au savoir de l'entreprise. Il n'est plus nécessaire de chercher dans les dossiers, les chats et les anciens e-mails. On peut poser une question en langage naturel et recevoir une réponse fondée sur les documents internes.

## Scène 8
La principale différence, c'est le contrôle. L'entreprise choisit les modèles, les sources de connaissances, les règles d'accès et le lieu de stockage des données. Ce n'est pas seulement un chatbot. C'est une fabrique locale d'assistants métier pour des tâches réelles.
