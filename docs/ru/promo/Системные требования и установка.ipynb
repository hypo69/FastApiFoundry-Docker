# Системные требования и установка

## Сцена 1
Локальный AI-агент запускается внутри вашей инфраструктуры. Его можно установить на рабочую станцию, локальный сервер или выделенную машину в периметре организации.

## Сцена 2
Минимальная конфигурация подходит для знакомства и небольших задач: современный процессор, 16 гигабайт оперативной памяти и достаточно места на диске для моделей, документов и журналов работы.

## Сцена 3
Для комфортной работы с большими документами, базой знаний и несколькими агентами лучше использовать 32 гигабайта памяти и видеокарту NVIDIA. Чем больше модель и объем данных, тем важнее запас по памяти и диску.

## Сцена 4
Основной вариант установки - Docker. Он упрощает запуск, изолирует компоненты и позволяет быстро повторить установку на другом компьютере или сервере.

## Сцена 5
Порядок установки простой. Сначала устанавливается Docker. Затем скачивается проект из GitHub. После этого настраивается файл конфигурации: где хранить данные, какие модели использовать, какие сервисы включить.

## Сцена 6
Далее система запускается одной командой. Контейнеры поднимают интерфейс, API, базу знаний и служебные компоненты. После запуска администратор открывает локальный адрес в браузере и проверяет состояние сервисов.

## Сцена 7
Следующий шаг - загрузить первые документы и создать базу знаний. Можно добавить регламенты, инструкции, таблицы, логи или тестовый набор файлов. Агент начнет отвечать с опорой на эти материалы.

## Сцена 8
После проверки можно подключать пользователей, задавать права доступа и создавать специализированных ассистентов под разные отделы. Установка остается локальной, а данные остаются под вашим контролем.
