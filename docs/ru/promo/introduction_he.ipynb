

# בינה מלאכותית מקומית שאפשר להפקיד בידיה נתונים אמיתיים.



# סצנה 1

כיום, אחד מכל שישה אנשים בעולם משתמש באופן פעיל בבינה מלאכותית גנרטיבית. ChatGPT, Nano Banana, Suno, Grok. מספר המודלים והדיוק שלהם גדלים בקצב מדהים.

# סצנה 2

הם נוחים. הם מהירים. הם עונים יפה. אבל בכל פעם אנחנו שולחים נתונים החוצה. זה יכול להיות מסמך. מייל. דוח. יומן אירועים. נוהל ייצור.

# סצנה 3

ומה לגבי אבטחה? לצערנו, כאן התמונה הרבה פחות ורודה. כל שאלה שנשאלת, כל קובץ שמועלה, נשלחים לענן. כך פועלים כל שירותי ה-AI הפופולריים. איפה נמצא הענן הזה, ומה קורה לבקשות שלנו? אני לא יודע. לכן לא אשתף מידע אישי עם גורם לא ידוע.

# סצנה 4

אז איך אפשר לשלב AI בעבודה היומיומית כאשר מסמכים משפטיים - חוזים, פסקי דין, התכתבויות, חוות דעת מקצועיות ומסמכים רפואיים -
מכילים מידע אישי וחוסים תחת סודיות?
כשאתם שולחים מסמך ל-AI בענן, אתם שולחים אותו לשרת של מישהו אחר. מבחינה משפטית, זו העברת מידע לצד שלישי. זה בלתי קביל לפי החוק, לפי האתיקה המקצועית ולפי ההיגיון הבריא.

# סצנה 5

שמנו את פרטיות הנתונים במרכז.

# סצנה 6

ChatGPT, Claude, Gemini
ושירותים נוספים הם שירותי ענן. אנחנו פותרים את הבעיה אחרת. אנחנו יוצרים עוזר מקומי שחי על המחשב שלכם או על שרת בתוך הארגון. המסמכים, הבקשות, התשובות, בסיס הידע, יומני העבודה והמידע התפעולי נשארים כולם בתוך גבולות העסק. אין צורך באינטרנט לצורך העבודה המרכזית. קוד הפרויקט פתוח לבדיקה. אתם יודעים בדיוק מה אתם מריצים, איך מוגדרת ההגנה, ואיפה נשמרים יומני העבודה. בלי תהליכים נסתרים.

# סצנה 7

בסיס ידע משלכם. פשוט מעלים טבלאות, נהלים פנימיים, פרוטוקולים של ישיבות או לוגים של שרתים. העוזר לא עונה בסיסמאות כלליות, אלא נשען בקפדנות על המסמכים שלכם. זהו פתרון מוכן למחלקות משפטיות, מרפאות, צוותי כספים וצוותי הנדסה. הנתונים נשארים במקום, ואיכות הניתוח עולה באופן חד. מפעל ליצירת סוכנים ועוזרים אישיים הוא ערכת כלים חזקה.

# סצנה 8


 אנחנו לא יוצרים עוד צ'אטבוט אופנתי. אנחנו נותנים לעסק שליטה מלאה. אתם בוחרים אילו נתונים לשמור בבסיס הידע, אתם בוחרים את המודלים, ואתם מגדירים את כללי הגישה. אתם מקבלים לא רק כלי עזר, אלא מפעל שלם לעוזרים מתמחים. אחד מנתח מסמכים, שני מנטר שרתים, שלישי עוקב אחרי קו הייצור. וכל זה נשען על תשתית מקומית אחת.

# סצנה 9

שאלה חשובה נוספת היא העלות. הקוד זמין באופן חופשי ב-GitHub. נשמח לקבל ספונסרים.


בינה מלאכותית הופכת לכלי עבודה אמיתי רק כאשר אפשר לסמוך עליה עם נתוני עבודה אמיתיים. ואפשר לסמוך רק על מה שאתם שולטים בו מתחילתו ועד סופו. ארכיטקטורה מקומית. קוד פתוח. הנתונים שלכם - הכללים שלכם.

נכון להיום, השלמנו שמונים אחוז מהקוד, שמונים אחוז מהבדיקות ושמונים אחוז מהתיעוד.
