# Prérequis système et installation

## Scène 1
L'agent d'IA local s'exécute à l'intérieur de votre infrastructure. Il peut être installé sur un poste de travail, un serveur local ou une machine dédiée dans le périmètre de l'organisation.

## Scène 2
La configuration minimale convient à la découverte et aux petites tâches : un processeur moderne, 16 gigaoctets de mémoire vive et suffisamment d'espace disque pour les modèles, les documents et les journaux de travail.

## Scène 3
Pour travailler confortablement avec de gros documents, une base de connaissances et plusieurs agents, il est préférable d'utiliser 32 gigaoctets de mémoire et une carte graphique NVIDIA. Plus le modèle et le volume de données sont importants, plus la réserve de mémoire et de disque devient essentielle.

## Scène 4
La méthode d'installation principale est Docker. Il simplifie le lancement, isole les composants et permet de reproduire rapidement l'installation sur un autre ordinateur ou serveur.

## Scène 5
Le processus d'installation est simple. On installe d'abord Docker. Ensuite, on télécharge le projet depuis GitHub. Puis on ajuste le fichier de configuration : où stocker les données, quels modèles utiliser et quels services activer.

## Scène 6
Le système est ensuite lancé avec une seule commande. Les conteneurs démarrent l'interface, l'API, la base de connaissances et les composants de service. Après le lancement, l'administrateur ouvre l'adresse locale dans le navigateur et vérifie l'état des services.

## Scène 7
L'étape suivante consiste à téléverser les premiers documents et à créer une base de connaissances. On peut ajouter des règlements, des consignes, des tableaux, des logs ou un jeu de fichiers de test. L'agent commence alors à répondre à partir de ces contenus.

## Scène 8
Après vérification, il est possible de connecter des utilisateurs, de définir les droits d'accès et de créer des assistants spécialisés pour différents services. L'installation reste locale, et les données restent sous votre contrôle.
