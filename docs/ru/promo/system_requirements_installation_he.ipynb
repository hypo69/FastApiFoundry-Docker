# דרישות מערכת והתקנה

## סצנה 1
סוכן ה-AI המקומי פועל בתוך התשתית שלכם. אפשר להתקין אותו על תחנת עבודה, על שרת מקומי או על מכונה ייעודית בתוך גבולות הארגון.

## סצנה 2
התצורה המינימלית מתאימה להיכרות ולמשימות קטנות: מעבד מודרני, 16 גיגה-בייט זיכרון RAM, ומספיק מקום בדיסק עבור מודלים, מסמכים ויומני עבודה.

## סצנה 3
לעבודה נוחה עם מסמכים גדולים, בסיס ידע וכמה סוכנים, עדיף להשתמש ב-32 גיגה-בייט זיכרון ובכרטיס מסך NVIDIA. ככל שהמודל ונפח הנתונים גדולים יותר, כך חשוב יותר להשאיר מרווח בזיכרון ובדיסק.

## סצנה 4
דרך ההתקנה המרכזית היא Docker. הוא מפשט את ההפעלה, מבודד רכיבים ומאפשר לשחזר את ההתקנה במהירות על מחשב או שרת אחר.

## סצנה 5
תהליך ההתקנה פשוט. קודם מתקינים Docker. לאחר מכן מורידים את הפרויקט מ-GitHub. אחר כך מגדירים את קובץ התצורה: היכן לשמור נתונים, באילו מודלים להשתמש ואילו שירותים להפעיל.

## סצנה 6
בשלב הבא מפעילים את המערכת בפקודה אחת. הקונטיינרים מעלים את הממשק, ה-API, בסיס הידע ורכיבי השירות. לאחר ההפעלה, מנהל המערכת פותח את הכתובת המקומית בדפדפן ובודק את מצב השירותים.

## סצנה 7
השלב הבא הוא להעלות את המסמכים הראשונים וליצור בסיס ידע. אפשר להוסיף נהלים, הוראות, טבלאות, לוגים או סט קבצים לבדיקה. הסוכן יתחיל לענות על בסיס החומרים האלה.

## סצנה 8
לאחר הבדיקה אפשר לחבר משתמשים, להגדיר הרשאות גישה וליצור עוזרים מתמחים למחלקות שונות. ההתקנה נשארת מקומית, והנתונים נשארים בשליטתכם.
