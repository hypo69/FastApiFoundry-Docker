# Requisitos del sistema e instalación

## Escena 1
El agente de IA local se ejecuta dentro de su infraestructura. Puede instalarse en una estación de trabajo, en un servidor local o en una máquina dedicada dentro del perímetro de la organización.

## Escena 2
La configuración mínima es adecuada para evaluación y tareas pequeñas: un procesador moderno, 16 gigabytes de RAM y suficiente espacio en disco para modelos, documentos y registros de trabajo.

## Escena 3
Para trabajar cómodamente con documentos grandes, una base de conocimiento y varios agentes, es mejor usar 32 gigabytes de memoria y una GPU NVIDIA. Cuanto más grande sea el modelo y el volumen de datos, más importante será contar con margen de memoria y disco.

## Escena 4
El método principal de instalación es Docker. Simplifica el arranque, aísla los componentes y permite repetir rápidamente la instalación en otro ordenador o servidor.

## Escena 5
El proceso de instalación es sencillo. Primero se instala Docker. Luego se descarga el proyecto desde GitHub. Después se ajusta el archivo de configuración: dónde almacenar los datos, qué modelos usar y qué servicios activar.

## Escena 6
A continuación, el sistema se inicia con un solo comando. Los contenedores levantan la interfaz, la API, la base de conocimiento y los componentes de servicio. Tras el arranque, el administrador abre la dirección local en el navegador y comprueba el estado de los servicios.

## Escena 7
El siguiente paso es cargar los primeros documentos y crear una base de conocimiento. Se pueden añadir reglamentos, instrucciones, hojas de cálculo, logs o un conjunto de archivos de prueba. El agente empezará a responder basándose en esos materiales.

## Escena 8
Después de la verificación, se pueden conectar usuarios, definir permisos de acceso y crear asistentes especializados para distintos departamentos. La instalación sigue siendo local, y los datos permanecen bajo su control.
