# System Requirements and Installation

## Scene 1
The local AI agent runs inside your infrastructure. It can be installed on a workstation, a local server, or a dedicated machine within the organization's perimeter.

## Scene 2
The minimum configuration is suitable for evaluation and small tasks: a modern processor, 16 gigabytes of RAM, and enough disk space for models, documents, and work logs.

## Scene 3
For comfortable work with large documents, a knowledge base, and several agents, it is better to use 32 gigabytes of memory and an NVIDIA GPU. The larger the model and the data volume, the more important memory and disk headroom become.

## Scene 4
The primary installation method is Docker. It simplifies launch, isolates components, and makes it easy to repeat the installation on another computer or server.

## Scene 5
The installation process is straightforward. First, Docker is installed. Then the project is downloaded from GitHub. After that, the configuration file is adjusted: where to store data, which models to use, and which services to enable.

## Scene 6
Next, the system is started with a single command. Containers bring up the interface, API, knowledge base, and service components. After launch, the administrator opens the local address in a browser and checks the service status.

## Scene 7
The next step is to upload the first documents and create a knowledge base. You can add policies, instructions, spreadsheets, logs, or a test set of files. The agent will begin answering based on these materials.

## Scene 8
After verification, you can connect users, define access rights, and create specialized assistants for different departments. The installation remains local, and the data remains under your control.
