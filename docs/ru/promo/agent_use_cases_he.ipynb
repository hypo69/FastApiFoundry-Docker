# תרחישי שימוש בסוכן

## סצנה 1
סוכן AI מקומי שימושי בכל מקום שבו עסק עובד עם מסמכים אמיתיים, תהליכים פנימיים ונתונים שאי אפשר פשוט לשלוח לענן.

## סצנה 2
מחלקה משפטית יכולה להעלות חוזים, פסקי דין, התכתבויות וחוות דעת מקצועיות. הסוכן עוזר לאתר סיכונים במהירות, להשוות גרסאות של מסמכים, להכין סיכומים קצרים ולענות על שאלות על בסיס חומרי התיק.

## סצנה 3
מרפאה יכולה להשתמש בסוכן לעבודה עם נהלים פנימיים, פרוטוקולים שעברו אנונימיזציה, הוראות וחומרי עזר. התנאי החשוב נשמר: הנתונים נשארים בתוך הארגון.

## סצנה 4
צוות כספים מקבל עוזר לניתוח דוחות, טבלאות, רשימות תשלומים וכללים פנימיים. הסוכן יכול להסביר חריגות, למצוא את המספרים הנכונים ולהכין טיוטה של הערה ניהולית.

## סצנה 5
צוותי הנדסה ו-IT יכולים לחבר לוגים של שרתים, מדריכים, תיעוד והיסטוריית תקלות. הסוכן עוזר להבין מה קרה, אילו מערכות הושפעו, ואילו צעדים כבר מתוארים בנהלים הפנימיים.

## סצנה 6
חברה יצרנית יכולה ליצור עוזרים עבור נהלים, כרטיסי ציוד, הוראות בטיחות ודוחות משמרת. סוכן אחד עובד עם תיעוד, אחר עוקב אחרי אירועים, ושלישי מסייע בהכשרת עובדים.

## סצנה 7
עבור מנהל, סוכן כזה הופך לנקודת גישה אחת לידע הארגוני. אין צורך לחפש בתיקיות, בצ'אטים ובמיילים ישנים. אפשר לשאול בשפה טבעית ולקבל תשובה שמבוססת על המסמכים הפנימיים.

## סצנה 8
ההבדל המרכזי הוא שליטה. הארגון בוחר את המודלים, את מקורות הידע, את כללי הגישה ואת מקום אחסון הנתונים. זה לא רק צ'אטבוט. זה מפעל מקומי לעוזרים עסקיים למשימות אמיתיות.
